```python
import os
import requests
from bs4 import BeautifulSoup

url = "https://ftp.gwdg.de/pub/misc/MPI-Leipzig_Mind-Brain-Body-LEMON/EEG_MPILMBB_LEMON/EEG_Preprocessed_BIDS_ID/EEG_Preprocessed/"

# CHANGE with your local directory where you want to save the files
local_dir = r"C:\Users\labo-maths\Downloads\eeg_oc"

os.makedirs(local_dir, exist_ok=True)

# Get the HTML page content
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Extract all file links
files = [a['href'] for a in soup.find_all('a') if not a['href'].endswith("/")]

# Download each file
for file in files:
    file_url = url + file
    local_file = os.path.join(local_dir, file)
    
    print(f"Downloading {file}...")
    file_response = requests.get(file_url, stream=True)
    with open(local_file, "wb") as f:
        for chunk in file_response.iter_content(chunk_size=8192):
            f.write(chunk)
    
    print(f"Downloaded: {file}")
```

```python
import os
import re
import pandas as pd

local_dir = r"C:\Users\labo-maths\Desktop\Babayan2018 FreSpeD\Data"

# Directory containing the files
directory = local_dir

# Regex pattern to extract the subject ID
pattern = re.compile(r"sub-(\d+)_")

# Set to store unique IDs
subject_ids = set()

# Loop through files in the directory
for filename in os.listdir(directory):
    match = pattern.search(filename)
    if match:
        subject_ids.add(match.group(1))

# Convert to DataFrame
df = pd.DataFrame(sorted(subject_ids), columns=["Subject_ID"], dtype=str)

df.head()
```
